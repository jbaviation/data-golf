```python
params = {'key': dg.api_key, 'file_format': 'csv', 'year': 2021}
endpt = 'https://feeds.datagolf.com/preds/pre-tournament-archive'

r = requests.get(endpt, params=params)
csv_list = [row.split(',') for row in r.content.decode("utf-8").split('\n')]
df = pd.DataFrame(csv_list[1:], columns=csv_list[0])
```
