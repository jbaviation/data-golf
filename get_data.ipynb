# Get Golf Data from datagolf.com
This notebook is created to help make functions that end up in `get_golf_data.py`. The functions should be tested here and then added to the *.py* file for production.

Getting data from [datagolf.com](https://datagolf.com/) will likely be more effective and consistent than building scrapers and crawlers from [pgatour.com](https://www.pgatour.com/).

```python
import pandas as pd
import numpy as np
import requests
import pygolf
from importlib import reload
import csv

_ = reload(pygolf)
```

```python
dg = pygolf.datagolf()

test = dg.get_tour_schedules()
```

```python
df = test.copy()

df['city'] = df['location'].str.extract(r'^([^,]*)')
df['us_state'] = df['location'].str.extract(r'([A-Z]{2})')
df['other_state'] = df['location'].str.extract(r'^[^,]*,([^,]*),')
df['country'] = df['location'].str.extract(r'([^,]*)$')

for col in ['city', 'us_state', 'other_state', 'country']:
    df[col] = df[col].str.strip()

df['country'] = np.where(df['us_state']==df['country'], 'United States', df['country'])
```
